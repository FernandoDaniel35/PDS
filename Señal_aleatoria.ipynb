```python
import numpy as np
import matplotlib.pyplot as plt

def generar_ruido_uniforme(n_muestras):
    """
    Genera una seal de ruido uniforme con varianza  = 4.
    """
    a = -2 * np.sqrt(3)
    b = 2 * np.sqrt(3)
    return np.random.uniform(a, b, n_muestras)

# Parmetros
n = 1000  # nmero de muestras

# Generar la seal
ruido = generar_ruido_uniforme(n)

# Calcular estadsticas
media = np.mean(ruido)
varianza = np.var(ruido)

# Graficar
plt.figure(figsize=(12, 4))
plt.plot(ruido, label='Ruido uniforme ( = 4)')
plt.axhline(media, color='red', linestyle='--', label=f'Media  {media:.2f}')
plt.title('Seal de Ruido Uniforme con Varianza  = 4')
plt.xlabel('Tiempo (ndice de muestra)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Mostrar resultados numricos
print(f"Media estimada: {media:.4f}")
print(f"Varianza estimada: {varianza:.4f}")
```

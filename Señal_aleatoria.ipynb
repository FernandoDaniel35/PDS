```python
import numpy as np
import matplotlib.pyplot as plt

def generar_ruido_uniforme(n_muestras):
    """
    Genera una señal de ruido uniforme con varianza σ² = 4.
    """
    a = -2 * np.sqrt(3)
    b = 2 * np.sqrt(3)
    return np.random.uniform(a, b, n_muestras)

# Parámetros
n = 1000  # número de muestras

# Generar la señal
ruido = generar_ruido_uniforme(n)

# Calcular estadísticas
media = np.mean(ruido)
varianza = np.var(ruido)

# Graficar
plt.figure(figsize=(12, 4))
plt.plot(ruido, label='Ruido uniforme (σ² = 4)')
plt.axhline(media, color='red', linestyle='--', label=f'Media ≈ {media:.2f}')
plt.title('Señal de Ruido Uniforme con Varianza σ² = 4')
plt.xlabel('Tiempo (índice de muestra)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Mostrar resultados numéricos
print(f"Media estimada: {media:.4f}")
print(f"Varianza estimada: {varianza:.4f}")
```

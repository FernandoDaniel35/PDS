# **Análisis de Densidad Espectral de Potencia (PSD) de la señal de audio "la cucaracha"**

### **1 Objetivo del Experimento**

El objetivo es estimar la PSD del audio con la musica de la cucaracha (cucaracha.wav) utilizando tres métodos (Periodograma, Welch, Blackman-Tukey) y, a partir de estas estimaciones, determinar el **ancho de banda** donde se concentra la **potencia dominante** de la señal.

### **1.1 Criterio de Ancho de Banda**

El BW se calcula como el rango de frecuencias donde la PSD supera un umbral estricto del **$5\%$ (0.05)** de su valor máximo. Este umbral asegura que solo se consideren las componentes de alta potencia, excluyendo el ruido y los armónicos muy débiles.

---

## **2\. Implementación de los Métodos de Estimación PSD**

Los tres métodos son **no paramétricos** y buscan estimar la PSD, pero difieren en el compromiso entre **Varianza** (ruido en la estimación) y **Resolución** (capacidad de distinguir frecuencias cercanas).

### **2.1 Método 1: Periodograma Ventaneado (Hamming)**

Es el método más directo: se calcula la **Transformada Rápida de Fourier (FFT)** de la señal completa y se eleva al cuadrado. Se aplica una **ventana** (Hamming en este caso) para reducir el **derrame espectral**.

$$\text{PSD}_{\text{Per}}(f) \propto \left| \text{FFT}\{x[n] \cdot w[n]\} \right|^2$$


#### **Implementación**

Se utiliza la señal completa ($N$ muestras) con una única ventana.

#### **Característica**

Ofrece la **máxima resolución** posible porque que utiliza toda la señal, pero tiene **alta varianza**. Esto resulta en una gráfica muy **dentada** o **ruidosa**.

### **2.2 Método 2: Método de Welch**

#### **Concepto y Proceso**

Método de **promediado**. Reduce la varianza (ruido) sacrificando un poco de resolución. Divide la señal en **segmentos más cortos y superpuestos**. Calcula el Periodograma de cada segmento y luego **promedia** los resultados.

$$\text{PSD}_{\text{Welch}}(f) = \text{Promedio} \left( \text{Periodograma} \{ \text{segmento}_i \} \right)$$


#### **Implementación**

Utiliza segmentos (ej: $2048$ muestras) con superposición (ej: $50\\%$ de solape) y una ventana (Hamming).

#### **Característica**

Proporciona una PSD **suave** y **clara** (baja varianza), lo que facilita el análisis visual de las principales frecuencias de potencia. Es el método más común para señales ruidosas.

### **2.3 Método 3: Blackman–Tukey**

#### **Concepto y Proceso**

Método **indirecto**. Se basa en el **Teorema de Wiener-Khinchin**, que establece que la PSD es la transformada de Fourier de la **Función de Autocorrelación (ACF)** de la señal. Se calcula la ACF, se le aplica una ventana (para suavizar) y se obtiene su FFT.


$${PSD}_{{BT}}(f) = {FFT} \left\{ R_{xx}[m] \cdot w[m] \right\}$$

Donde $R_{xx}[m]$ es la ACF y $w[m]$ es la ventana (Hamming).

#### **Implementación**

Calcula la ACF hasta un retardo máximo $M$ (ej: $1024$) y luego aplica la FFT.

#### **Característica**

Al igual que Welch, produce una PSD **suave** (baja varianza). La resolución está limitada por el valor de $M$.

## **3\. Resultados y Comparación del Ancho de Banda**

Los resultados del cálculo de ancho de banda, utilizando el **5% del máximo** como umbral, son los siguientes:

| Método | Frecuencia Baja (fbaja​) \[Hz\] | Frecuencia Alta (falta​) \[Hz\] | Ancho de Banda (BW) \[Hz\] |
| :---- | :---- | :---- | :---- |
| **Periodograma** | $1005.000$ | $2017.667$ | ${1012.667}$ |
| **Welch** | $1007.813$ | $1453.125$ | ${445.313}$ |
| **Blackman–Tukey** | $996.094$ | $1476.563$ | ${480.469}$ |

### **3.1 Análisis de Comparación de Resultados**

1. **Concentración de Potencia (${f\_{{baja}}}$):**  
   * Los tres métodos coinciden en que la potencia dominante ($\>5\%$ del máximo) comienza alrededor de **$1000 { Hz}$ ($1 { kHz}$)**. Esto indica que la **frecuencia fundamental** principal del audio se encuentra en el rango medio.  
2. **Discrepancia en Ancho de Banda (${BW}$):**  
   * **Welch y Blackman-Tukey** (Baja Varianza): Producen un BW estrecho y consistente de aproximadamente **$450 { Hz}$** ($≈ 1450 { Hz} - 1000 { Hz}$). Estos métodos, al promediar o suavizar, eliminan el ruido aleatorio y los armónicos débiles. Por lo tanto, solo las frecuencias que tienen una **potencia estable y real** logran superar el umbral del $5\%$.  
   * **Periodograma** (Alta Varianza): Arroja un BW mucho mayor: **$1012 { Hz}$**. El Periodograma no promedia, lo que permite que el **ruido aleatorio** se sume a picos armónicos débiles. Estos picos de ruido, aumentados por la alta varianza, extienden el espectro hasta los $2017 { Hz}$ y superan el umbral del $5\%$ de forma espuria.

---

## **4\. Conclusiones**

El experimento demuestra que la elección del estimador de PSD es crítica, especialmente cuando se aplica un **umbral estricto** para definir una métrica como el ancho de banda.

| Conclusión | Periodograma | Welch / Blackman-Tukey |
| :---- | :---- | :---- |
| **Varianza (Ruido)** | Muy Alta | Muy Baja |
| **BW Estimado al 5%** | **Sobreestimado** ($≈ 1012 { Hz}$) | **Realista** ($≈ 450 { Hz}$) |
| **Motivo** | El ruido es aumentado por la alta varianza y **supera artificialmente** el umbral del $5\%$. | El ruido es **atenuado** por el promediado/suavizado, revelando el verdadero núcleo de potencia. |

A los fines prácticos, en el análisis de señales de audio, para determinar la banda de energía principal, los métodos de **baja varianza** como **Welch** y **Blackman-Tukey** son superiores. El umbral de 5% es demasiado sensible al ruido para ser útil con un umbral estricto.

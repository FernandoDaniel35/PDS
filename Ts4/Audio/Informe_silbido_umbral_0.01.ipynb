# **Análisis Espectral de Potencia (PSD) de la señal de Silbido Tonal elevando el umbral a 10% del máximo**

## **1\. Análisis y Objetivo**

### **1.1 Modificación del Umbral y Propósito**

El análisis se realizó sobre el archivo silbido.wav, que contiene un silbido humano. El código utilizado, Silbido\_umbral\_0.1.py, ajusta el cálculo del Ancho de Banda (BW) con un **umbral relativo del ${10\%}$ (${0.1}$)**.

El propósito de aumentar el umbral (anteriormente $1\%$) fue doble:

1. **Excluir el ruido de fondo de banda ancha** que dominaba los resultados anteriores.  
2. **Aislar las componentes de alta potencia** del tono puro, incluyendo armónicos, para obtener una estimación más precisa de la región energética del silbido.

### **1.2 Métodos de Estimación PSD**

Se emplearon tres métodos no paramétricos para estimar la PSD y calcular el BW:

* **Periodograma ventaneado**  
* **Método de Welch**  
* **Blackman-Tukey**

---

## **2\. Resultados de Ancho de Banda (Umbral 10%)**

La siguiente tabla resume los resultados obtenidos al aplicar el umbral del $10\%$ a la PSD del silbido:

| Método | Frecuencia Baja (fbaja​) \[Hz\] | Frecuencia Alta (falta​) \[Hz\] | Ancho de Banda (BW) \[Hz\] |
| :---- | :---- | :---- | :---- |
| **Periodograma** | $2678.333333$ | $6622.333333$ | ${3944.0000}$ |
| **Welch** | $2671.875000$ | $6632.812500$ | ${3960.9375}$ |
| **Blackman–Tukey** | $2660.156250$ | $6644.531250$ | ${3984.3750}$ |

---

## **3\. Análisis**

Los resultados muestran una **alta consistencia** entre los tres métodos, con un BW promedio de ${≈ 3960 { Hz}}$ (${≈ 4 { kHz}}$). Esta banda abarca desde ${≈ 2.7 { kHz}}$ hasta ${≈ 6.6 { kHz}}$.

### **3.1 Éxito en la Exclusión del Ruido de Baja Frecuencia**

El aumento del umbral fue exitoso en acotar el límite inferior:

* En el análisis anterior (umbral $1\%$), la ${f\_{{baja}}}$ se ubicaba en $0 { Hz}$ o cerca, indicando que el ruido de baja frecuencia y la componente DC superaban el umbral.  
* En este análisis (umbral $10\%$), la ${f\_{{baja}}}$ se establece consistentemente en ${≈ 2.7 { kHz}}$. Esto confirma que el umbral del $10\%$ logró **excluir el ruido de fondo** y la potencia DC, aislando la región donde la energía es *realmente* significativa (más del 10% del máximo).

### **3.2 Naturaleza de la Señal del Silbido**

El BW resultante de $≈ 4 { kHz}$ es todavía muy amplio para un tono puro ideal (que debería ser de unas pocas decenas o cientos de Hz). Esto lleva a la conclusión de que la señal del silbido.wav es un tono que:

* **Posee Armónicos o Ruido de Banda Ancha Fuerte:** La potencia superior al $10\%$ se extiende desde ${2.7 { kHz}}$ hasta ${6.6 { kHz}}$. Esto sugiere que el silbido es rico en **armónicos** o que el tono principal tiene un **ruido intrínseco de banda lateral** muy fuerte, y estas componentes también superan el umbral de $10\%$.  
* **Frecuencia Alta Consistente:** La ${f\_{{alta}}}$ en ${≈ 6.6 { kHz}}$ en los tres métodos demuestra que las componentes de potencia relevante terminan de forma abrupta en esa región.

### **3.3 Estabilidad de los Estimadores**

La diferencia en el BW entre el resultado más bajo (Periodograma: $3944.00 { Hz}$) y el más alto (Blackman-Tukey: $3984.38 { Hz}$) es de solo $≈ 40 { Hz}$, lo que representa una **variación inferior al ${1\%}$** entre los métodos. Esta alta concordancia indica que los tres estimadores están midiendo con precisión la misma región espectral de alta potencia.

---

## **4\. Conclusiones**

| Estimador | Frecuencia Mínima con Potencia \> 10% | BW Reportado |
| :---- | :---- | :---- |
| **Periodograma** | ${2.68 { kHz}}$ | ${3944.00 { Hz}}$ |
| **Welch** | ${2.67 { kHz}}$ | ${3960.94 { Hz}}$ |
| **Blackman-Tukey** | ${2.66 { kHz}}$ | ${3984.38 { Hz}}$ |

El umbral de ${10\%}$ fue una mejora al eliminar el ruido de baja frecuencia, revelando que la energía significativa del silbido.wav se distribuye a lo largo de una banda de **aproximadamente ${4 { kHz}}$** en la región de ${2.7 { kHz}}$ a ${6.6 { kHz}}$.

# **Análisis Espectral de Potencia (PSD) de la Señal de Silbido**

## **1\.Objetivo**

### **1.1 Naturaleza del Silbido en el Espectro**

El silbido humano, por su naturaleza, produce un **tono casi puro**, concentrando la mayor parte de su energía en un **pico espectral muy estrecho** (generalmente entre $1 { kHz}$ y $5 { kHz}$). La Densidad Espectral de Potencia (PSD) debería reflejar esta concentración de energía en un BW idealmente muy reducido.

### **1.2 Objetivo del Experimento**

Estimar la PSD del archivo silbido.wav con los métodos de **Periodograma ventaneado**, **Welch** y **Blackman-Tukey** para evaluar cómo la concentración de potencia de un tono puro interactúa con el ruido de fondo cuando se utiliza un umbral de potencia muy bajo.

### **1.3 Criterio de Ancho de Banda**

El ancho de banda se calcula como el rango de frecuencias donde la PSD supera el umbral del **${1\% (0.01)}$** de su valor máximo. Los resultados indican que este umbral es excesivamente bajo, capturando el **ruido de banda ancha** de la grabación a lo largo de casi todo el espectro disponible.

## **3\. Análisis y Discusión de los Resultados**

Los anchos de banda (BW) de ${≈ 6.7 { kHz}}$ contradicen el BW estrecho esperado para un tono puro. Este fenómeno ocurre porque el **umbral del 1% está capturando el ruido de banda ancha** de la grabación en lugar de la energía específica del silbido.

### **3.1 Frecuencia Alta y Ancho de Banda (${BW}$)**

* **Dominio del Ruido:** Los tres métodos reportan una **frecuencia alta muy cercana a ${6.8 { kHz}}$**, lo que resulta en BWs prácticamente idénticos (entre $6.68 { kHz}$ y $6.82 { kHz}$).  
* **Significado:** Esta uniformidad sugiere que el **ruido de fondo** es lo suficientemente potente para superar el umbral del $1\%$ a lo largo de casi toda la banda de frecuencias analizada. En estas condiciones, el BW medido está determinado por la banda de ruido del sistema y no por la característica tonal del silbido.

### **3.2 Frecuencia Baja (${f_{{baja}}}$)**

* **Periodograma y Blackman-Tukey** estiman ${f_{{baja}}} = 0 { Hz}$. Esto implica que el ruido de muy baja frecuencia o la **potencia DC** lograron superar el umbral establecido.  
* **Welch** estima ${f_{{baja}}} = 117.19 { Hz}$. La técnica de promediado de segmentos de Welch es más efectiva para mitigar la influencia de la potencia DC, resultando en un corte inferior que excluye la frecuencia cero.

---

## **4\. Conclusiones**

El análisis demuestra que en el caso de un **tono puro con ruido de fondo**, la elección del **umbral** es más crítica que la elección del estimador PSD.

| :---- | :---- | :---- | :---- |
| **BW Dominante** | Ruido de Banda Ancha | Ruido de Banda Ancha | Ruido de Banda Ancha |

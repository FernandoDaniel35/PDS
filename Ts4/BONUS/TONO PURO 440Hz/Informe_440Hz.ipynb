# **Análisis Espectral de Potencia (PSD) de Tono Puro (440 Hz)**

## **1\. Análisis y Objetivo**

### **1.1 Naturaleza de la Señal y Umbral**

El archivo de audio 440Hz\_44100Hz\_16bit\_05sec.wav es un **tono puro** de ${440 { Hz}}$ (nota 'La'), que es una señal **determinística**. En teoría, su Densidad Espectral de Potencia (PSD) debe ser un impulso con un Ancho de Banda (BW) cercano a cero.

El análisis utiliza un **umbral relativo del ${5\%}$ (${0.05}$)** de la PSD máxima para medir la dispersión de energía alrededor del pico principal.

### **1.2 Métodos de Estimación PSD**

Se compararon tres métodos que difieren en su compromiso entre **Resolución** y **Varianza**:

* **Periodograma ventaneado:** Prioriza la **máxima resolución**.  
* **Método de Welch:** Prioriza la **baja varianza** (suavizado espectral por promediado).  
* **Blackman-Tukey:** Prioriza la **baja varianza** (suavizado por ventaneado de la Autocorrelación).

## **3\. Análisis**

### **3.1 El Periodograma: Precisión para Tonos Puros**

El **Periodograma** ofrece el resultado **más preciso** para la medición de un tono puro:

* **BW de ${0.40 { Hz}}$:** Este valor está muy cerca del BW ideal nulo.  
* **Fundamento:** La **alta resolución** del Periodograma es más efectivo para distinguir el pico de energía estrecho. La pequeña dispersión se debe principalmente al **efecto de la ventana de tiempo** (pérdida espectral) sobre la señal de duración finita, no a la varianza de la estimación.

### **3.2 Welch y Blackman-Tukey:**

Los métodos de baja varianza **sobreestiman** el ancho de banda por un factor de hasta $160$:

* **BW de ${21.5 { Hz}}$ (Welch) y ${64.6 { Hz}}$ (Blackman-Tukey):** Estos BW son demasiado amplios.  
* **Fundamento:** La reducción de la varianza se logra mediante el **suavizado espectral** (promediado/ventaneado). En una señal tonal, este suavizado **ensancha el pico espectral** original. El umbral del $5\%$ se alcanza en frecuencias más lejanas a $440 { Hz}$, aumentando el BW calculado.  
* **Blackman-Tukey** muestra la **mayor sobreestimación** ($≈ 64.6 { Hz}$), lo que indica que el ventaneado de la autocorrelación impone el mayor ensanchamiento espectral de los tres métodos.

### **3.3 Frecuencia Central**

A pesar de la gran diferencia en el BW, los tres métodos confirman que el **centro de la potencia** de la señal es la frecuencia fundamental de ${440 { Hz}}$, ya que las frecuencias límite (${f\_{{baja}}}$ y ${f\_{{alta}}}$) son simétricas alrededor de ese valor.

## **4\. Conclusiones**

| Estimador | Principal Compromiso | BW Obtenido | Adecuación para Tono Puro |
| :---- | :---- | :---- | :---- |
| **Periodograma** | Alta Resolución | ${0.40 { Hz}}$ (Preciso) | **El más recomendado**. Su resolución es necesaria para medir el BW muy estrecho de señales determinísticas. |
| **Welch** | Baja Varianza (Suavizado) | ${21.53 { Hz}}$ (Sobreestimado) | **No recomendado**. El suavizado distorsiona la medición del BW. |
| **Blackman-Tukey** | Baja Varianza (Suavizado) | ${64.60 { Hz}}$ (Muy Pobre) | **No recomendado**. Introduce el mayor ensanchamiento espectral. |

**Conclusión:** Para el análisis de **señales determinísticas de tonos puros**, el **Periodograma** es el estimador superior. Los métodos de baja varianza (Welch y Blackman-Tukey), muy buenos para señales ruidosas y de banda ancha como por ejemplo la voz, resultan de menor resolución, lo que distorsiona la medición del ancho de banda de tonos puros.

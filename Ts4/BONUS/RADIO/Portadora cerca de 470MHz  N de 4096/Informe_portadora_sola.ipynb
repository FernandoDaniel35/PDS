# **ANÁLISIS ESPECTRAL de una PORTADORA PURA sin MODULAR**

## **1\. Introducción y Objetivos del Experimento**

---

Este informe presenta el análisis espectral de una **portadora de radiofrecuencia (${RF}$)** capturada por un SDR, cuyo principal objetivo era verificar su **estado no modulado** y medir su mínima ocupación de ancho de banda.

### **1.1. Contexto de la Señal**

La señal de interés fue capturada cerca de los $470 { MHz}$ y almacenada en el archivo ${IQ.csv}$. Esta señal es una **portadora continua (CW) o portadora pura**, sin información.

* **Frecuencia:** $470.000 { MHz}$.  
* **Ancho de Banda de Cálculo:** $±100 kHz$ (un canal de $200 { kHz}$).

### **1.2. Objetivos Principales**

1. **Cuantificar el Ancho de Banda Ocupado (${BW}$):** Determinar la mínima ocupación espectral de una señal pura.  
2. **Verificar el Estado de Modulación:** Confirmar que la señal no contiene información (BW cercano a cero).  
3. **Evaluar Métodos ${PSD}$:** Comparar la precisión de los métodos de estimación espectral para un pico de frecuencia.

## **2\. Resultados del Análisis Espectral**

---

El cálculo del Ancho de Banda (${BW}$) se realizó encontrando el rango donde la Densidad Espectral de Potencia (${PSD}$) supera el **$5\%$** del máximo dentro del límite de $±100 { kHz}$.

### **Ancho de Banda Estimado (${BW}$)**

| Método | Frecuencia Baja \[MHz\] | Frecuencia Alta \[MHz\] | Ancho de banda \[kHz\] |
| :---- | :---- | :---- | :---- |
| Periodograma | $469.990000$ | $469.990000$ | ${0.000000}$ |
| Welch | $469.988000$ | $469.992000$ | ${4.000000}$ |
| Blackman–Tukey | $469.988995$ | $469.990996$ | ${2.000977}$ |

### **Observaciones**

El ancho de banda medido es **extremadamente pequeño** en todos los casos, confirmando el estado no modulado de la señal. Los valores son mucho menores que los $≈ 67 { kHz}$ observados en la señal modulada del experimento con la portadora modulada en FM banda ancha.

---

## **3\. Análisis**

### **3.1. Interpretación del Ancho de Banda Mínimo**

El resultado indica que el ${BW}$ medido es **casi cero** (${0 }a{ 4 kHz}$).

* **Portadora Pura:** En el mundo ideal, una portadora de frecuencia única (una onda sinusoidal) tiene toda su energía concentrada en un solo punto del espectro, lo que significa que su ancho de banda es teóricamente **cero**.  
* **Mundo Real:** el pequeño ancho de banda observado ($2 { kHz}$ o $4 { kHz}$) no es modulación. Es el resultado inevitable de:  
  1. **Ruido de Fase:** Imperfecciones en los osciladores del receptor (${SDR}$), que hacen que el pico puro se **ensanche** ligeramente.  
  2. **Ruido de Cuantificación:** El ruido inherente al proceso de digitalización (muestreo y redondeo de valores).

El ${BW}$ es tan pequeño que se encuentra completamente dentro del **ruido de la medición**, no de una señal de información.

### **3.2. Frecuencia Real de la Portadora**

Los resultados indican que el pico de potencia no está en $470.000 { MHz}$, sino concentrado alrededor de ${469.990 { MHz}}$.

* **Conclusión:** La frecuencia real de la portadora está ${10 { kHz}}$ **por debajo** de la frecuencia de sintonización objetivo ($470.000 { MHz}$). Esta diferencia es una pequeña **desviación de frecuencia** o un error de sintonía al momento de capturar los datos.
Se debe a las partes por millón de error en la frecuencia a la que oscila el cristal de cuarzo de referencia del PLL del receptor SDR.
---

### **4. Mejor Método:**

Para esta tarea específica, la medición de un **pico único y agudo** (la portadora pura), la elección del mejor método se invierte respecto al análisis de la señal modulada.

| Método | Resultado BW | Varianza (Ruido) | Mejor para... |
| :---- | :---- | :---- | :---- |
| **Periodograma** | ${0.00 { kHz}}$ | Alta | Medir la **máxima resolución** o picos puros. |
| Welch | $4.00 { kHz}$ | Baja | Medir la potencia total o ${BW}$ de **señales complejas** (con bandas laterales). |
| Blackman–Tukey | $2.00 { kHz}$ | Baja | Similar a Welch. |

### **El mejor método resultó ser: El Periodograma**

El **Periodograma** demostró ser el más preciso para esta medición:

* **Razón:** El Periodograma ofrece la mejor **resolución espectral** posible. Fue capaz de aislar el pico puro de la portadora tan limpiamente que el umbral del $5\%$ se cumplió solo en el punto máximo, dando el ${BW}$ ideal de ${0 { kHz}}$.  
* **Problema con Welch y ${BT}$:** Los métodos Welch y Blackman-Tukey, al **promediar** para reducir el ruido, involuntariamente **ensanchan** el pico puro de la portadora. Esto provocó que el umbral del $5\%$ se activara en un rango de $2 { a } 4 { kHz}$, dando una lectura de ${BW}$ **falsamente más alta** de lo que la portadora debería ocupar.

## **5\. Conclusiones**

1. **Verificación de Señal Pura:** La señal es una portadora no modulada, confirmada por su Ancho de Banda Ocupado de ${0 { kHz}}$ (según la medición más precisa).  
2. **Desviación Confirmada:** La portadora se localiza con precisión en ${469.990 { MHz}}$.  
3. Se demostró que la elección del estimador de ${PSD}$ es definitorio para efectuar una medición correcta:  
   * Para medir el ${BW}$ de señales complejas (como la ${FM}$ anterior), **Welch** es superior.  
   * Para medir la resolución y el ${BW}$ de picos puros (como esta portadora), el **Periodograma** es el mejor.

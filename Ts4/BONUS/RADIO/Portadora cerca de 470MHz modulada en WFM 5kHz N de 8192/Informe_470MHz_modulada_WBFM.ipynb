# **ANÁLISIS ESPECTRAL y de MODULACIÓN de SEÑAL de 470MHz**

## **1\. Introducción y Objetivos del Experimento**

### **1.1. Contexto de la Señal**

El análisis se centró en una **señal de radiofrecuencia (${RF}$)** capturada por un receptor de radio definido por *software* (${SDR}$, modelo ${RTL}2832$) y guardada como datos $I/Q$ en el archivo ${IQ.csv}$.

* **Frecuencia Central (Portadora):** $470 { MHz}$.  
* **Tipo de Modulación Esperada:** Frecuencia Modulada de Banda Ancha (${WBFM}$).  
* **Señal Moduladora (Contenido):** Un tono puro sinusoidal de **$5 { kHz}$**.  
* **Ancho de Banda del Canal Asignado:** $200 { kHz}$ ($±100 { kHz}$ alrededor de la portadora).

### **1.2. Objetivos**

El experimento buscó caracterizar la señal y validar las herramientas de procesamiento digital:

1. **Medir el Ancho de Banda Ocupado (${BW}$):** Determinar cuánto espacio del espectro utiliza realmente la señal.  
2. **Determinar la Desviación de Frecuencia ($∆f$):** Calcular la intensidad de la modulación ${FM}$ a partir del ${BW}$ medido.  
3. **Evaluar Métodos ${PSD}$:** Comparar tres técnicas de cálculo de densidad espectral de potencia para identificar la más robusta.

---

## **2\. Fundamentos de la Señal y Metodología (${I/Q}$)**

### **2.1. ¿Qué son los datos $I/Q$?**

Cuando un ${SDR}$ captura una señal de radio, la transforma en dos componentes separadas para el análisis digital:

* **Componente I (In-phase / En Fase):** La parte de la señal que está "alineada" con la portadora.  
* **Componente Q (Quadrature / En Cuadratura):** La parte de la señal que está desfasada $90°$ respecto a $I$.

Al combinar $I$ y $Q$, se forma la **Señal Compleja**. Esta señal permite a la computadora saber en todo momento no solo la **Amplitud** sino también la **Fase y Frecuencia** de la señal. Esto es fundamental, porque en **Modulación de Frecuencia (${FM}$)**, la información (el tono de $5 { kHz}$) se codifica en los **cambios de frecuencia**, no en la amplitud.

### **2.2. El Procesamiento: Densidad Espectral de Potencia (${PSD}$)**

La **${PSD}$** es la herramienta utilizada. Indica cómo se distribuye la energía total de la señal a lo largo de las frecuencias. Usando el ${PSD}$ podemos "ver" las bandas laterales del tono de $5 { kHz}$ y dónde termina la energía de la señal (su ${BW}$).

Se definió el ${BW}$ como el rango de frecuencias donde la potencia supera un **umbral del $5\%$** de la potencia máxima del espectro, limitado a un rango de $±100 { kHz}$ (el ancho de canal completo).

## **3\. Resultados del Análisis Espectral**

El código calculó la **${PSD}$** utilizando los tres métodos de ${PSD}$ (Periodograma, Welch, y Blackman-Tukey) para medir el Ancho de Banda Ocupado:

### **Ancho de Banda Estimado (${BW}$)**

| Método | Frecuencia Baja \[MHz\] | Frecuencia Alta \[MHz\] | Ancho de banda \[kHz\] |
| :---- | :---- | :---- | :---- |
| Periodograma | $469.957750$ | $470.024500$ | $66.750000$ |
| Welch | $469.956000$ | $470.024000$ | $68.000000$ |
| Blackman–Tukey | $469.956979$ | $470.024012$ | $67.032731$ |
| **Promedio** | \- | \- | **$67.26 { kHz}$** |

### **Conclusión**

Los resultados son consistentes, dando un Ancho de Banda promedio de **$67.26 { kHz}$**. Este valor es **mucho menor** que el ancho de canal disponible de $200 { kHz}$.

---

## **4\. Análisis: ¿Por qué $67 { kHz}$?**

El valor de ${BW} ≈ 67 { kHz}$ no es un error, es el reflejo de la modulación real.

### **4.1. La Regla de Carson**

Para una señal ${FM}$ modulada por un tono ($fm$), el ${BW}$ se rige por la Regla de Carson, la cual estima cuánto se extiende el espectro:

$${BW} = 2 ({∆f} + {fm})$$  
Donde $\\mathbf{∆f}$ (Desviación de Frecuencia Máxima) es el parámetro que ajusta el transmisor para hacer que la portadora se desvíe.

### **4.2. Cálculo de la Desviación Real ($∆f$)**

Utilizando el ${BW}$ medido ($≈ 67.2 { kHz}$) y sabiendo que $fm = 5 { kHz}$:

$$67.2 { kHz} ≈ 2 (∆f + 5 { kHz}) = 33.6 { kHz} ≈ ∆f + 5 { kHz} ⇒ {∆f ≈ 27.6 { kHz}}$$  
**Conclusión:** La señal capturada no estaba utilizando la desviación máxima típica de $75 { kHz}$ (que daría un ${BW}$ de $\\approx 152 { kHz}$), sino una desviación de solo **$27.6 { kHz}$**. Esto confirma que el código está midiendo el **espectro real ocupado** por la señal bajo las condiciones de modulación que se estaban usando en ese momento.

**Nota:**
Con un equipo HP8920 medí la desviación real, la frecuencia de la portadora y la frecuencia de la señal modulante. Observar la foto incluída en la carpeta. Se concluye que la medición con el SDR y el cálculo se aproximan mucho a la realidad. No es posible llegar a una estimación exacta del BW porque el SDR posee varios factores que hacen que no sea tan preciso como un instrumento HP8920 de Hewlett Packard.

## **5\. Evaluación de los Métodos de ${PSD}$**

Los tres métodos funcionan, pero tienen diferencias:

### **Periodograma**

* El gráfico es muy **"puntiagudo"** y con picos.  
* **Problema:** Tiene **alta varianza** (mucho ruido), lo que hace que la curva sea inestable. Al buscar el ${BW}$ con un umbral del $5\%$, el ruido puede hacer que la curva cruce el umbral incorrectamente.

### **Blackman-Tukey**

* El gráfico es **suave**, similar a Welch.  
* **Problema:** Es más complejo de implementar y su estimación puede ser un poco más sesgada que Welch.

### **Método de Welch**

* Divide la señal en pequeños bloques, calcula el espectro de cada bloque y luego **promedia** los resultados.  
* **Ventaja:** Reduce drásticamente la **varianza** (ruido), produciendo una curva ${PSD}$ **suave y estable**.  
* **Conclusión:** Para la tarea de **medir el Ancho de Banda** utilizando un umbral de potencia (como el $5\%$), el **Método de Welch** es el más robusto, ya que su suavidad asegura que la medición del umbral no se vea afectada por picos de ruido momentáneos.

## **6\. Conclusiones**

1. **La Medición es Válida:** El Ancho de Banda Ocupado fue medido con alta precisión en ${67.26 { kHz}}$, demostrando que la señal está mucho más concentrada de lo que el ancho de canal de $200 { kHz}$ sugiere.  
2. **$∆f$ Real Conocida:** Se ha determinado que el transmisor estaba modulando con una **Desviación Máxima de $≈ 27.6 { kHz}$** para el tono de $5 { kHz}$.  En una medición con HP8920(ver foto en la carpeta) se observó que la desviación real era de aproximadamente **30kHz**.

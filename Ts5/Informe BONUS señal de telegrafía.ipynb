## I. Comparativa

La principal diferencia observada entre los filtros de Respuesta de Impulso Finita (**FIR**) y los de Respuesta de Impulso Infinita (**IIR**) radica en su comportamiento de fase en la Banda de Paso (**BP**).

1.  **Filtros FIR (Kaiser y Parks-McClellan):**
    * Mantienen una **Fase Lineal** excelente.
    * El **Retardo de Grupo (GD)** es **constante** en aproximadamente $\mathbf{1999.5}$ **muestras** para todos los anchos de banda.
    * Resultado: **Mínima Distorsión de Fase** en la señal filtrada.

2.  **Filtros IIR (Chebyshev I y Cauer/Elíptico):**
    * Presentan una **Fase No Lineal**.
    * El **Retardo de Grupo** es **no constante** y se vuelve extremadamente variable (picos altos) en los bordes de la BP a medida que se reduce el ancho de banda.
    * Resultado: Introducen **Distorsión de Fase** que altera la forma de la señal en el dominio del tiempo.

## II. Retardo de Grupo y Distorsión de Fase

### Retardo de Grupo (Group Delay)

El **Retardo de Grupo ($\tau_g$)** es una métrica crucial que mide el tiempo de retraso (o retardo) que experimenta cada componente de frecuencia al pasar por el filtro. Se define matemáticamente como la derivada negativa de la respuesta de fase, $\tau_g(\omega) = -d\phi(\omega)/d\omega$.

* **Fase Lineal:** Ocurre cuando el Retardo de Grupo es constante en la Banda de Paso. Si todas las frecuencias se retrasan la misma cantidad, la forma de la señal se mantiene intacta. Los filtros FIR, al ser simétricos, aseguran esta característica (Retardo Constante), lo que garantiza la **mínima distorsión temporal**.
* **Fase No Lineal (Distorsión de Fase):** Ocurre cuando el Retardo de Grupo varía con la frecuencia (curva). Esto significa que las componentes de frecuencia se desalinean temporalmente, **distorsionando** la forma de la señal filtrada.

### Significado de los Picos Bruscos ("Rayas Verticales") en el Retardo de Grupo

Los picos grandes y bruscos, vistos en los gráficos como "rayas verticales", son la manifestación visual de la **Máxima Distorsión de Fase localizada** en los filtros IIR.

#### 1. Pendiente Abrupta
La fórmula $\tau_g(\omega) = -d\phi(\omega)/d\omega$ establece que el Retardo de Grupo es el negativo de la **pendiente** de la curva de fase.

* **El Retardo Extremo:** Un pico muy alto y estrecho (una "raya vertical") en el gráfico de $\tau_g$ implica que la pendiente de la fase ($d\phi/d\omega$) es **extremadamente negativa y muy pronunciada** en ese punto de frecuencia específico.
* **El Efecto de las Rayas:** Estas rayas verticales aparecen porque la fase experimenta un cambio **abrupto, no suave**, en un intervalo de frecuencia muy pequeño (es decir, $\Delta \omega$ es minúsculo). Al calcular la derivada, este cambio abrupto se traduce en un valor instantáneo de $\tau_g$ gigantesco, que visualmente se asemeja a un impulso o una "raya" vertical en lugar de una curva.

#### 2. Los Polos
El fenómeno está asociado al diseño de los filtros IIR para cumplir con la alta selectividad ($\alpha_s=80 { dB}$) con un orden bajo.

* **Polos cerca de la Banda de Paso:** Para maximizar la selectividad (transición estrecha), los **polos** del filtro deben estar ubicados **muy cerca del círculo unitario** y, por lo tanto, muy cerca de las frecuencias de corte de la Banda de Paso.
* **Efecto en la Fase:** La proximidad extrema de un polo provoca que la contribución de fase cambie **drásticamente** a medida que la frecuencia de análisis se acerca a él.
* **Consecuencia Temporal:** Estas frecuencias se retrasan una cantidad desproporcionadamente larga (cientos o miles de muestras). Por ejemplo, en el filtro Cauer/Elíptico con ${BW}=100 { Hz}$, los picos alcanzan $\approx 2500$ muestras.

En resumen, las **"rayas verticales"** no son ruido, sino **puntos de máxima distorsión de fase**, donde el filtro está forzando al máximo sus características de selectividad, resultando en un retardo temporal localizado y extremo.

## IV. Conclusiones Generales

1.  **FIR:** Los filtros FIR (Kaiser y Parks-McClellan) demuestran ser extremadamente lineales en su respuesta de fase, manteniendo un Retardo de Grupo constante (Fase Lineal) de $1999.5$ muestras en todos los anchos de banda, independientemente de la mayor selectividad requerida. Esto los hace ideales si la **preservación temporal** de la señal es la prioridad.
2.  **IIR:** Los filtros IIR (Chebyshev I y Cauer/Elíptico) logran el mismo objetivo de atenuación ($\mathbf{80 { dB}}$) con órdenes de filtro mucho menores. Sin embargo, este beneficio se logra a costa de la **Fase Lineal**, introduciendo una distorsión no lineal que se manifiesta como picos altos en el Retardo de Grupo.
3.  **Impacto del Ancho de Banda:** A medida que el ancho de banda se estrecha (de $500 { Hz}$ a $100 { Hz}$), la **distorsión de fase de los IIR se incrementa drásticamente**, ya que los picos de retardo de grupo alcanzan valores más altos y más dispersos (hasta $\approx 2500$ muestras en BW=$100 { Hz}$). Esto es un efecto directo de la necesidad de acercar los polos al círculo unitario para lograr mayor selectividad con un orden bajo.

***

## V. Conclusiones Finales: Recepción del tono , Eficiencia y Costo computacional

Este análisis final permite establecer un claro compromiso entre las familias de filtros FIR e IIR basado en los requerimientos del sistema de recepción CW:

1.  **Mejor Rendimiento para Recepción (Integridad Temporal / Inocuidad):**
    * **Mejor de todos:** Filtros **FIR (Kaiser y Parks-McClellan)**.
    * **Justificación:** Los FIR son los únicos que ofrecen **Fase Lineal Perfecta** (retardo de grupo constante de $\approx 1999.5$ muestras), garantizando la **mínima o nula distorsión temporal** en el pulso CW. Esto es crucial en sistemas de comunicación donde la forma la señal debe ser preservada para una correcta decodificación. Los IIR introducen una distorsión que se agrava al estrechar el ancho de banda (BW=100 Hz, picos de GD de $\approx 2500$ muestras), volviéndolos **no inocuos** para la integridad de la señal.

2.  **Mejor Eficiencia y Menor Costo Computacional:**
    * **Mejor de todos:** Filtros **IIR (Chebyshev I y Cauer/Elíptico)**.
    * **Justificación:** Los IIR logran el requisito de atenuación extrema ($\mathbf{80 { dB}}$) con un **orden de filtro muchísimo menor** que el orden fijo de $N=4000$ utilizado para los FIR. Un orden menor implica un menor número de coeficientes de filtro (*taps*) y, por lo tanto, una **extrema reducción en las operaciones de multiplicación y suma** por muestra, lo que se traduce en un **mayor rendimiento computacional** y menor latencia en aplicaciones en tiempo real.
                      

3.  **Conclusión Final:**
    * La elección depende de la prioridad del diseño: Si la **integridad de la señal y la linealidad de fase** son críticas, los **FIR** son indispensables, asumiendo un **alto costo computacional** (orden $N=4000$). Si el **rendimiento computacional** y el orden bajo son la prioridad, los **IIR** son la mejor opción, pero a costa de introducir una **distorsión de fase elevada** que impacta en la forma de la señal.

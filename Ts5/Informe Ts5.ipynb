## I. Diseño de Filtros IIR (Chebyshev I y Cauer/Elíptico)

El objetivo es diseñar dos filtros IIR (Chebyshev Tipo I y Cauer/Elíptico) para una señal de ECG muestreada a $f_s = 1000$ Hz, cumpliendo una plantilla estricta de rechazo a baja y alta frecuencia (ruido de línea base y ruido muscular/electrodos).

### 1. Especificaciones de Diseño (Comunes a IIR y FIR)

| Parámetro | Valor | Descripción |
| :--- | :--- | :--- |
| Frecuencia de Muestreo ($f_s$) | 1000 Hz | |
| Frecuencia de Paso (**$w_p$**) | (0.8 Hz, 35 Hz) | |
| Frecuencia de Detención (**$w_s$**) | (0.1 Hz, 40 Hz) | |
| Atenuación Máx. en Banda de Paso ($\alpha_p$) | 1 dB | |
| **Atenuación Mín. en Banda de Detención ($\mathbf{\alpha_s}$)** | **40 dB** | **Criterio de diseño** |

### 2. Diseño y Resultados de Filtros IIR

El diseño se realizó utilizando la función `sig.iirdesign` de `scipy`, obteniendo los siguientes resultados:

| Filtro | Tipo de Respuesta | Orden (N) | Observaciones |
| :--- | :--- | :--- | :--- |
| **Chebyshev Tipo I** | Equirriple en banda de paso | **23** | Cumple la atenuación de 40 dB solo en los puntos críticos. |
| **Cauer / Elíptico** | Equirriple en ambas bandas | **12** | **Orden mínimo** para $40$ dB. **Máxima selectividad**. |

**Conclusión del Diseño IIR:** El filtro **Cauer/Elíptico** es **más eficiente**, ya que logra la misma atenuación mínima ($\alpha_s = 40$ dB) con un orden **casi dos veces menor** ($N=12$ vs $N=23$), garantizando la **máxima selectividad** gracias a la presencia de ceros de transmisión en las bandas de detención.

### 3. Análisis de Polos y Ceros (Plano Z)

Los filtros IIR (Respuesta al Impulso Infinita) son definidos por su función de transferencia $H(z)$ como el cociente de dos polinomios:

$$H(z) = \frac{B(z)}{A(z)} = \frac{b_0 + b_1 z^{-1} + \cdots + b_N z^{-N}}{a_0 + a_1 z^{-1} + \cdots + a_N z^{-N}}$$

Los **ceros** son las raíces de $B(z)$ y los **polos** son las raíces de $A(z)$.

| Elemento | Conceptual | Posición y Consecuencia |
| :--- | :--- | :--- |
| **Polos ($z_{polo}$)** | Definen la **estabilidad** y la **pendiente de la transición**. | Para la **estabilidad** de un sistema LTI discreto, **todos los polos deben estar estrictamente dentro del círculo unitario** en el plano Z ($\mathbf{\lvert z_{polo} \rvert < 1}$). Su cercanía al círculo unitario aumenta la selectividad (transiciones más abruptas) a costa de la introducción de rizos en la banda de paso (Chebyshev I). |
| **Ceros ($z_{cero}$)** | Definen los puntos de **máxima atenuación** o **rechazo de banda**. | La ubicación de un cero **exactamente sobre el círculo unitario** ($\mathbf{\lvert z_{cero} \rvert = 1}$) produce una **atenuación infinita** (un 'notch' o cero de transmisión) en la frecuencia angular correspondiente. |

#### 1. Filtro IIR Chebyshev I (N=23)

El filtro Chebyshev I es un diseño de **rizo en la banda de paso**. Su diagrama de Polos y Ceros es el siguiente:



* **Polos (Marcas 'x'):** Los polos se encuentran agrupados **cerca del círculo unitario**, concentrados en la región de la **banda de paso** (cerca de $z=e^{j\omega_p T}$). Esta cercanía al círculo unitario es la causa directa del **rizado (ripple)** en la banda de paso, ofreciendo una transición rápida hacia la banda de detención. La posición de los polos determina la frecuencia de resonancia y la selectividad del filtro.
* **Ceros (Círculos 'o'):** El filtro Chebyshev I no introduce ceros en el círculo unitario para forzar la atenuación a un valor específico. Los ceros visibles, como el par en $z=\pm 1$ (frecuencia DC y Nyquist), son intrínsecos al mapeo de la transformación bilineal a un filtro pasa-banda.
* **Estabilidad:** El gráfico muestra que todos los polos están **dentro** del círculo unitario ($|p_i| < 1$), garantizando la **estabilidad** del filtro.

#### 2. Filtro IIR Cauer/Elíptico (N=12)

El filtro Cauer/Elíptico, siendo el más eficiente, presenta rizado tanto en la banda de paso como en la banda de detención.



* **Polos (Marcas 'x'):** Al igual que el Chebyshev I, los polos están cerca del círculo unitario, pero la distribución es **más concentrada** y simétrica. Esto, junto con el efecto de los ceros, permite que el filtro Elíptico cumpla las mismas especificaciones que el Chebyshev I con un **orden significativamente menor (N=12 vs N=23)**, demostrando su mayor eficiencia.
* **Ceros (Círculos 'o'):** El Cauer/Elíptico es el único IIR que introduce **ceros en el círculo unitario** (en la banda de detención, $z=e^{j\omega_s T}$). Estos ceros de transmisión (visibles en las proximidades de las frecuencias de corte $\omega_s=0.1$ Hz y $\omega_s=40$ Hz) son los responsables de los **picos de atenuación** y del **rizado** en la banda de detención, permitiendo la caída más abrupta posible.
* **Estabilidad:** De igual manera, todos los polos están **dentro** del círculo unitario, confirmando su **estabilidad**.

La mayor selectividad y eficiencia del filtro Cauer/Elíptico ($N=12$) se explica directamente por la ubicación estratégica de sus ceros finitos sobre el círculo unitario, creando puntos de atenuación infinita en las bandas de rechazo.

Este gráfico (referencia) muestra que ambos filtros IIR (Chebyshev I y Cauer/Elíptico) cumplen visualmente con los 40 dB de atenuación.

Este zoom de la banda de transición de baja frecuencia (referencia - Zoom 0.65 a 0.9 Hz) ilustra la **superior selectividad del filtro Cauer/Elíptico**.

* La caída del Cauer/Elíptico ($N=12$, púrpura) es **más abrupta** y alcanza la atenuación de $-40$ dB antes que el Chebyshev I ($N=23$, naranja). Esto es crucial para la **eliminación de la deriva de la línea base**.

Este zoom de la banda de transición de alta frecuencia (referencia - Zoom 33 a 41 Hz) muestra que ambos diseños alcanzan y superan los $40$ dB de atenuación cerca de la banda de detención superior (40 Hz).

## II. Diseño de Filtros FIR (Boxcar y Parks-McClellan)

El diseño de los filtros FIR se basa en la misma plantilla, pero se utiliza un orden arbitrariamente alto ($N=3000$) para forzar el cumplimiento de la atenuación de $40$ dB. Los filtros FIR tienen la ventaja de ser de **fase lineal** (no introducen distorsión de fase), pero requieren un **orden mucho mayor** que los IIR.

### 1. Parámetros de Diseño FIR

| Parámetro | Valor | Fuente/Función |
| :--- | :--- | :--- |
| Orden del Filtro (N) | **3000** | Fijo para intentar cumplir especificación |
| Frecuencia de Paso ($w_p$) | (0.8 Hz, 35.0 Hz) | |
| Frecuencia de Detención (estimada) | (0.1 Hz, 35.7 Hz) | **Predistorsión** (para Remez) |
| Atenuación Mín. ($\alpha_s$) | **40 dB** | Criterio de diseño |

### 2. Diseño y Resultados de Filtros FIR

| Filtro | Tipo de Ventana / Algoritmo | Orden (N) | Observaciones |
| :--- | :--- | :--- | :--- |
| **Boxcar (Ventana Rectangular)** | Ventana simple (Función `firwin2`) | **3000** | **Fracaso en la atenuación de 40 dB** debido al **Fenómeno de Gibbs**. |
| **Parks-McClellan (Remez)** | Algoritmo Equirriple (`remez`) | **3000** | **Cumple la atenuación de 40 dB** en frecuencia, a costa de un orden masivo. |

### 3. Análisis de Polos y Ceros (Plano Z) y Fase Lineal

Los filtros FIR (Respuesta al Impulso Finita) son definidos por su función de transferencia $H(z)$ como un polinomio en $z^{-1}$:

$$H(z) = b_0 + b_1 z^{-1} + \cdots + b_N z^{-N} = \frac{B(z)}{1}$$

| Elemento | Conceptual | Posición y Consecuencia |
| :--- | :--- | :--- |
| **Polos ($z_{polo}$)** | Definen la **estabilidad**. | El denominador de $H(z)$ es $1$. Esto implica que **todos los polos de un filtro FIR se encuentran en el origen** del plano Z ($\mathbf{z_{polo} = 0}$). Conceptualmente y matemáticamente, esto asegura que el sistema es **estable** y **no recursivo**. |
| **Ceros ($z_{cero}$)** | Definen la **respuesta en frecuencia**. | Un filtro FIR de orden $N$ tiene exactamente $N$ ceros. La ubicación de estos ceros sobre o cerca del **círculo unitario** determina la respuesta en frecuencia. Los ceros de un filtro pasa-banda se agrupan: $\mathbf{N_{cero}}$ ceros cerca de $\mathbf{z=1}$ (DC, para rechazar bajas frecuencias) y $\mathbf{N_{cero}}$ ceros cerca de $\mathbf{z=-1}$ (Nyquist, para rechazar altas frecuencias). |

#### 1. Filtro FIR Boxcar (N=3000)

El filtro diseñado con la Ventana Rectangular (Boxcar) tiene un orden muy alto ($N=3000$) y presenta la siguiente configuración de Polos y Ceros:



* **Polos (Marcas 'x'):** Una característica fundamental de **todos los filtros FIR** es que sus **polos se encuentran exclusivamente en el origen** del Plano Z, es decir, en $z=0$. Esto implica que la función de transferencia $H(z)$ solo tiene términos en el numerador, lo que garantiza la **estabilidad** del sistema.
* **Ceros (Círculos 'o'):** Los $N$ ceros determinan la respuesta en magnitud del filtro. En el caso del filtro Boxcar, los ceros están distribuidos alrededor del círculo unitario. La falta de optimización de la ventana rectangular resulta en lóbulos laterales grandes, que se reflejan en la densidad y posición de estos ceros, especialmente en las zonas de transición y detención.
* **Fase Lineal:** La simetría compleja de los ceros con respecto al eje real y la circunferencia unitaria **garantiza la fase lineal**, lo que implica un **Retardo de Grupo Constante** (igual a $N/2 = 1500$ muestras) y nula distorsión de fase para la señal de interés.

#### 2. Filtro FIR Remez/Parks-McClellan (N=3000)

El filtro diseñado mediante el algoritmo equirriple (Parks-McClellan/Remez) ofrece la mínima duración de impulso (orden) para unas especificaciones dadas. Su diagrama es:



* **Polos (Marcas 'x'):** Al ser un filtro FIR, sus polos también se encuentran **únicamente en el origen ($z=0$)**, asegurando la estabilidad.
* **Ceros (Círculos 'o'):** El algoritmo Remez coloca los ceros estratégicamente **exactamente sobre el círculo unitario** ($|z|=1$), particularmente en las **bandas de detención** ($0$ Hz y $40$ Hz). Esta colocación precisa es la que produce la característica **equirriple** (ripples uniformes) en la banda de detención y permite la transición más rápida posible entre bandas con el orden $N$ fijado.
* **Fase Lineal:** Este filtro también exhibe una **simetría perfecta de los ceros** en el Plano Z, lo que, como en todos los FIR, se traduce en una **fase lineal** y un **retardo de grupo constante** ($\tau_g = N/2$) en todas las frecuencias.

### 4. Gráficos de Respuesta en Frecuencia FIR

#### a) Detalle de Magnitud (0 Hz a 1 Hz)



Este zoom (referencia - Detalle 0Hz a 1Hz) muestra:

* **FIR Boxcar (azul):** No alcanza los $-40$ dB cerca de $0.1$ Hz. Confirma el **Fenómeno de Gibbs** y el **incumplimiento de la especificación**.
* **FIR Remez (rojo):** Alcanza y cumple los $-40$ dB en la banda de detención de baja frecuencia (0.1 Hz).

#### b) Fase



Este gráfico (referencia - Fase) demuestra la principal ventaja de los FIR: ambos presentan una **fase perfectamente lineal** en la banda de paso, lo que significa que **no introducen distorsión de fase** (retardo de grupo constante).

## Análisis Comparativo de Filtros Pasa-Banda

El objetivo de esta sección es verificar el rendimiento de los cuatro diseños de filtros en el **dominio del tiempo** sobre una señal de ECG fuertemente contaminada, analizando 13 zonas de interés predefinidas en el script `Ts5.py`.

### Introducción y Metodología

Los cuatro filtros comparados, diseñados para cumplir la plantilla de especificaciones de atenuación mínima de $\alpha_s = 40$ dB en bandas de detención, son:

1. **IIR Chebyshev I** (Orden $N=23$,)
2. **IIR Cauer/Elíptico** (Orden $N=12$,)
3. **FIR Boxcar (Ventana Rectangular)** (Orden $N=3000$,)
4. **FIR Parks-McClellan (Remez)** (Orden $N=3000$,)

**Metodología de Evaluación:**
* **Inocuidad (Fase Cero):** Verificación de la preservación de la morfología P-QRS-T mediante la aplicación bidireccional (`filtfilt`/`sosfiltfilt`), eliminando el retardo de fase.
* **Eficiencia:** Capacidad de **corregir la deriva de la línea base** (ruido de baja frecuencia) y **suavizar** el ruido de alta frecuencia, en cumplimiento de la atenuación de $40$ dB en ambas bandas de detención.

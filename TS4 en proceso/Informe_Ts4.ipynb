
## Informe del Análisis Espectral de Señales realizado

El presente informe detalla el análisis efectuado, aplicando de los tres métodos de Estimación de la Densidad Espectral de Potencia (DEP). Se analizan las señales de **Audio**, **Biomédicas** y de **Radiofrecuencia (RF)**, presentes en los archivos de muestras recibidos.

### 1. Comparación de los métodos de Estimación de la Densidad Espectral de Potencia (DEP)

El objetivo de la DEP es determinar **cómo se distribuye la energía** de una señal en función de la **frecuencia**. La elección del método es un compromiso entre la **resolución** (capacidad de distinguir frecuencias cercanas) y la **varianza** (la confiabilidad de la estimación de potencia).

| Método | Procedimiento | Objetivo | Ventaja |
| :--- | :--- | :--- | :--- |
| **Periodograma Ventaneado** | FFT de la señal completa con aplicación de ventana (e.g., Hamming). | Análisis Rápido, Máxima **Resolución** Espectral. | Muestra detalles finos (ruido y picos) pero es **inconsistente** (alta varianza). |
| **Método de Welch** | Promedio de múltiples Periodogramas de segmentos superpuestos. | Obtener una estimación **Consistente** (baja varianza) y robusta. | Es el estándar para señales que **varían** (voz, música, RF), porque **promedia las fluctuaciones**. |
| **Blackman-Tukey** | FFT de la Función de **Autocorrelación** de la señal. | Obtener una curva DEP **Suave** para el análisis de la forma espectral. | Excelente para visualizar la envolvente de potencia; útil cuando la **varianza** debe ser mínima. |

### 2. Análisis de Señales de Audio

Los scripts (`Silbido.py`, `La_cucaracha.py`, `prueba.py`) analizan el contenido frecuencial en el espectro audible.

#### Objetivos y Conclusiones:

| Señal | Fenómeno Físico | Resultado Esperado | Mejor método |
| :--- | :--- | :--- | :--- |
| **Silbido** | **Tono puro**. | Un único pico de alta energía. | **Blackman-Tukey**. Suaviza el espectro al máximo, aislando el tono fundamental del silbido de cualquier ruido de fondo, permitiendo una medición estable. |
| **Voz** | **Cuasi-periódica** (armónicos) modulada $\sim200-4000\text{ Hz}$). | Espectro con banda de energía ancha y decaimiento rápido. | **Método de Welch**. La voz es **no estacionaria**. El promedio es crucial para obtener la forma suave de la envolvente del espectro. |
| **La cucaracha** | Frecuencia fundamental **cambia** con el tiempo (notas musicales). | Múltiples picos que representan las notas musicales usadas. | **Método de Welch**. Promedia la energía de todas las notas. El ancho de banda resultante será el rango total de frecuencias cubiertas por la canción. |

### 3. Análisis de Señales Biomédicas

Los scripts (`Ts4_ECG.py`, `Ts4_Pletismografia.py`) se enfocan en las señales del corazón (ECG) y el flujo sanguíneo (PPG) en la banda de **muy baja frecuencia**, buscando la Frecuencia Cardíaca ($\mathbf{F_C}$) y la Frecuencia Respiratoria ($\mathbf{F_R}$).

#### Objetivos y Conclusiones:

| Señal | Fenómeno Físico | Frecuencias Clave | Mejor método |
| :--- | :--- | :--- | :--- |
| **ECG** (`ECG_TP4.mat`) | Actividad eléctrica. | $F_C$ (Frecuencia Cardíaca, $\sim1-2\text{ Hz}$) y sus armónicos. | **Método de Welch**. Reduce la **varianza** del ruido muscular y de línea base, haciendo que el pico de la $F_C$ sea más **confiable y estable** para diagnóstico. |
| **PPG** (`PPG.csv`) | Variaciones en el volumen sanguíneo. | $F_C$ y $F_R$ (Frecuencia Respiratoria, $\sim0.15-0.3\text{ Hz}$). | **Blackman-Tukey**. Su **suavidad** permite distinguir los picos de $F_C$ y $F_R$ que están muy próximos en el espectro. |

### 4. Análisis de Señales de Radiofrecuencia (RF) - FM (Rock and Pop)

El archivo `Rock and Pop.py` analiza una señal WFM (Modulación de Frecuencia de Banda Ancha) capturada en $\mathbf{95.9\text{ MHz}}$ por un SDR RTL-2832 y una placa BeagleBone Black.

#### Relación entre Ancho de Modulación y Ancho del Canal

1.  **Ancho de Modulación (Ancho de Banda de la Señal):** El ancho de banda real ocupado por la información (audio y subportadoras estéreo).
2.  **Ancho de Canal (Ancho Asignado):** El espacio total de frecuencia asignado a la emisora, que incluye las bandas de guarda.

* **Ancho de Modulación por norma:**
    * La norma para la FM de Banda Ancha establece que la **señal modulada** ocupa $\mathbf{\sim75\text{ kHz}}$ (desviación máxima de $\pm75\text{ kHz}$).
    * Este valor de $\mathbf{75\text{ kHz}}$ es **correcto** para transmisiones de **FM Estéreo** (que incluyen la subportadora de $38\text{ kHz}$), siendo de $\sim50\text{ kHz}$ para señales mono. Dado que la **Rock and Pop transmite en estéreo**, el resultado próximo a $\mathbf{\sim75\text{ kHz}}$ es **físicamente correcto** porque tomo la parte positiva del espectro, con lo cual la frecuencia de la portadora se traslada a 0Hz y lo que estoy midiendo es la parte superior del espectro a partir de la portadora.

* **Ancho de Canal por norma:**
    * La norma le asigna un **ancho de canal de $\mathbf{200\text{ kHz}}$** a la emisora. La diferencia entre $200\text{ kHz}$ y $75\text{ kHz}$ son las **bandas de guarda** requeridas para minimizar la interferencia con los canales adyacentes.

#### El Efecto de la Muestra Instantánea (Periodograma)

El archivo de datos (`Rock and Pop.csv`) contiene la **magnitud de una única FFT** (un Periodograma) de la señal modulada.

* **Validación de $\mathbf{75\text{ kHz}}$:** El hecho de que la aplicación del cálculo de ancho de banda (con el umbral del $5\%$) dé un valor cercano a $\mathbf{75\text{ kHz}}$ en esta **muestra instantánea** es un resultado muy **significativo y correcto**. Indica que, a pesar de la alta varianza y el ruido inherente a un Periodograma, la potencia de la portadora de WFM (FM de Banda Ancha)está predominantemente confinada al ancho que impone la modulación real.

* **Promediado (Método de Welch):**
    * Aunque el Periodograma $\mathbf{instantáneo}$ ya arroja un valor correcto (los $75\text{ kHz}$ de modulación), la forma del espectro es **ruidosa** y **no consistente** (alta varianza).
    * Para obtener una estimación **estable y suavizada** de esos mismos $75\text{ kHz}$, es imprescindible utilizar el **Método de Welch**.
    * **Welch** promediaría las fluctuaciones de potencia de la música a lo largo del tiempo, entregando la **curva de potencia más estable y confiable** para la modulación WFM de $\mathbf{75\text{ kHz}}$.

**Conclusión para RF:** La obtención de un ancho de banda cercano a $\mathbf{75\text{ kHz}}$ es **correcta** para el ancho de la señal WFM Estéreo. El script aplica los 3 métodos a una **muestra instantánea**, lo que demuestra que incluso bajo las limitaciones de una alta varianza, la potencia está concentrada en el ancho de modulación exigido por la norma. El **Método de Welch** seguiría siendo la herramienta preferida para garantizar la **consistencia** y la suavidad de la medición de los $\mathbf{75\text{ kHz}}$.

---

### 5. Conclusiones

1.  **Consistencia vs. Resolución:** El Periodograma se usa para la **máxima resolución** (detalles finos) y Welch/Blackman-Tukey para la **máxima consistencia** (confiabilidad de potencia).
2.  **Adecuación del Método:** Para las señales más complejas y variables (Voz, Música, ECG, RF), el **Método de Welch** es la herramienta más robusta, ya que su núcleo es el **promediado**, atenuando el efecto de las fluctuaciones de potencia y ruido que hacen que el Periodograma sea inutilizable en la práctica.
3.  **Ancho de Banda:** La función de ancho de banda con umbral (`threshold=0.05`) es un paso analítico crítico, forzando la interpretación a considerar solo el **rango de frecuencias que portan potencia significativa** (al menos el $5\%$ del máximo), lo cual es esencial para caracterizar la ocupación espectral.
